```python
df.isnull().sum()
```

# Machine Learning

```python
X = df[df.columns[2:]]
y = df["diagnosis"]
```

```python
X.head()
```

```python
y.head()
```

### Scaling the Numeric Attributes

```python
from scipy import stats
```

```python
X = X.apply(stats.zscore)
```

```python
X.head()
```

```python
# Alternate Method
X1 = df[df.columns[2:]]
X1 = (X1 - X1.mean()) / X1.std()
X1.head()
```

### Splitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### Applying kNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kNN_model = KNeighborsClassifier(n_neighbors=5, weights="distance")
```

```python
kNN_model.fit(X_train, y_train)
```

```python
y_pred = kNN_model.predict(X_test)
```

```python
y_pred[:5]
```

# Evaluating the performance of kNN model

```python
kNN_model.score(X_test, y_test)
```

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, y_pred, labels=["M", "B"])
cm
```

```python
cm_display = metrics.ConfusionMatrixDisplay(cm, display_labels=["M", "B"])
cm_display.plot()
```

```python
print("Classification Report: \n")
print(metrics.classification_report(y_test, y_pred, labels=["M", "B"]))
```

### Chosing k-value

```python
scores = {}
for k in range(1, 50):
    kNN = KNeighborsClassifier(n_neighbors=k, weights="distance")
    kNN.fit(X_train, y_train)
    scores[k] = kNN.score(X_test, y_test)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(scores.keys(), scores.values())
plt.xlabel("Value of k")
plt.ylabel("Score of model")
```

```python
sorted(scores.items(), key=lambda x: x[1], reverse=True)[:10]
```
